```python
import numpy as np

data = np.load("../data/raw/TimesNet-datasets/SMD/SMD_train.npy")
data.shape
```

```python
import pandas as pd

df = pd.DataFrame(data)
df.head()
```

```python
df.describe()
```

```python
import plotly.graph_objects as go

downsample_factor = 1
df_downsampled = df.iloc[::downsample_factor]

fig = go.Figure()
for col in df_downsampled.columns:
    fig.add_trace(
        go.Scatter(
            x=df_downsampled.index,
            y=df_downsampled[col],
            mode='lines',
            name=col
        )
    )
index_label = "Datetime" if isinstance(df_downsampled.index, pd.DatetimeIndex) else "Index"
fig.update_layout(
    title=f"SMD | original shape: {df.shape} | plot downsample factor: {downsample_factor}",
    xaxis_title=index_label,
    yaxis_title="Value",
    legend=dict(
        title="Columns",
        orientation="v",
        x=1.02,
        xanchor="left",
        y=1
    )
)
fig.show()
```

```python
df.to_csv("../data/processed/SMD_train.csv", index=False)
```
